```python
import pandas as pd

match_stats = pd.read_csv('epl_matches_stats.csv', index_col=0)
player_list = pd.read_csv('player_list_per_full.csv', index_col=0)

merged_stats = player_list.merge(match_stats, how = 'inner', on = ['date', 'team', 'opponent'])
```
